```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import re
from nltk.corpus import stopwords
import nltk
import tensorflow as tf
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = 'Natural language processing (NLP) is a field of computer science, artificial intelligence and computational linguistics concerned with the interactions between computers and human (natural) languages, and, in particular, concerned with programming computers to fruitfully process large natural language corpora. Challenges in natural language processing frequently involve natural language understanding, natural language generation (frequently from formal, machine-readable logical forms), connecting language and machine perception, managing human-computer dialog systems, or some combination thereof.'
```

```python
text=re.sub(r'[^a-zA-Z\s]','',text)
text=text.lower()
text
```

```python
tokenizer=tf.keras.preprocessing.text.Tokenizer()
text_sent_tokenised=nltk.sent_tokenize(text)
tokenizer.fit_on_texts(text_sent_tokenised)
```

```python
keys=tokenizer.word_index.keys()
list(keys)[:13]
```

```python
word_tokens=nltk.word_tokenize(text)
```

```python
word_tokens
```

```python
stop_words=set(stopwords.words('english'))
```

```python
stop_words
```

```python
corpus=[]
for w in word_tokens:
    if w not in stop_words:
        corpus.append(w)
```

```python
corpus
```

```python
from textblob import TextBlob, Word
```

```python
corpus_lemmanised=[Word(i) for i in corpus]
```

```python
corpus_lemmanised
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
```

```python
corpus_stemmed=[stemmer.stem(i) for i in corpus]
```

```python
corpus_stemmed
```

```python
from gensim.models import FastText
model_fast=FastText(vector_size=30, window=3, min_count=1)
```

```python
model_fast.build_vocab(corpus_iterable=corpus)
model_fast.train(corpus_iterable=corpus, total_examples=len(corpus),epochs=10)
```

```python
embeddings=[]
for i in corpus_stemmed:
    embeddings.append(model_fast.wv[i])
```
